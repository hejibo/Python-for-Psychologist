# Custom iPython HTML Formatters

```python
from IPython.display import display
ip = get_ipython()

MAX_ROW_INSTANCE_COUNT=50
```

###ioDataStore SessionMetaDataInstance objects

```python
def smd2html(smd):
    ntuple_class_template="""
    <li>
        <strong>
                Sessions   
        </strong>
    <ul>
    <span style="color:#006400;">
    <strong>
            Session Info:
    </strong>
    </span>
    """
    ntuple_fields_template="""
    <ul>
        {attributes}

    """  
    
    ntuple_attribute_template="""
    <li>
        <strong> {key}: </strong> {value}
    </li>
    """
    user_variables_template="""
    <li>
        <strong>
                User Variables
        </strong>
        <ul>
            {attributes}
        </ul>
    </li>
    </ul>
    </li></ul>
    """

    result = ntuple_class_template.format(class_name='SessionMetaDataInstance')
    attributes=''.join([ntuple_attribute_template.format(key=key,value=value)for key,value in smd._asdict().iteritems() if key !='user_variables'])
    result+=ntuple_fields_template.format(attributes=attributes)
    user_vars=''
    for ukey,uvar in smd.user_variables.iteritems():
       user_vars+=ntuple_attribute_template.format(key=ukey,value=uvar)
    result+=user_variables_template.format(attributes=user_vars)
    return result.replace('\\n','')


ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.datastore.util',
                                 'SessionMetaDataInstance', smd2html)
```

###ioDataStore ExperimentMetaDataInstance objects

```python

def emd2html(emd):
    ntuple_class_template="""
    <span style="font-family:arial,helvetica,sans-serif;">
    <ul>
    <span style="color:#800000;">
    <strong>
            Experiment Info:
    </strong>
    </span>
    """
    
    ntuple_fields_template="""
    <ul>
        {attributes}
    """  
    
    ntuple_attribute_template="""
    <li>
        <strong> {key}: </strong> {value}
    </li>
    """
    
    experiment_end_template="""
    </ul></ul>
    </span>
    """
    
    result = ntuple_class_template
    attributes=''.join([ntuple_attribute_template.format(key=key,value=value)for key,value in emd._asdict().iteritems() if key !='sessions'])
    result+=ntuple_fields_template.format(attributes=attributes)
    for smd in emd.sessions:
       result+=smd2html(smd)
    result+=experiment_end_template
    return result.replace('\\n','')

ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.datastore.util',
                                 'ExperimentMetaDataInstance', emd2html)
```

###list objects

```python
def list2Table(dlist):
    header_template="""
		<table align="left" border="1" cellpadding="1" cellspacing="1" style="width: 80%;">
			<thead>
				<tr>
                    {header_row_cells}
				</tr>
			</thead>
			<tbody>

    """

    header_row_cell_template="""
					<th scope="col">
						{header_cell_value}
                    </th>
    """
    
    column_template="""
					<td>
						{cell_value}
                    </td>
    """
    
    row_template="""
				<tr>
                    {columns}
				</tr>
    """
    
    table_end_template="""
			</tbody>
		</table>
    """
    
    h_row=dlist[0]
    hrow_cells=''.join([header_row_cell_template.format(header_cell_value=v) for v in h_row])
    result=header_template.format(header_row_cells=hrow_cells)
    for r in dlist[1:]:
        cells=''
        for c in r:
            cells+=column_template.format(cell_value=c)
        result+=row_template.format(columns=cells)
    result+=table_end_template
    return result

ip.display_formatter.formatters['text/html'].for_type(list,list2Table)
```

###PyTables Row objects

```python
def rowIterator2Table(row):
    header_template="""
		<table align="left" border="1" cellpadding="1" cellspacing="1" style="width: 80%;">
			<thead>
				<tr>
                    {header_row_cells}
				</tr>
			</thead>
			<tbody>

    """

    header_row_cell_template="""
					<th scope="col">
						{header_cell_value}
                    </th>
    """
    
    column_template="""
					<td>
						{cell_value}
                    </td>
    """
    
    row_template="""
				<tr>
                    {columns}
				</tr>
    """
    
    table_end_template="""
			</tbody>
		</table>
    """

    h_row=row.table.colnames
    hrow_cells=''.join([header_row_cell_template.format(header_cell_value=v) for v in h_row])
    result=header_template.format(header_row_cells=hrow_cells)
    try:
        rc=0
        while row:
            cells=''
            for c in h_row:
                cells+=column_template.format(cell_value=row[c])
            result+=row_template.format(columns=cells)
            row=row.next()
            rc+=1
            if rc>= MAX_ROW_INSTANCE_COUNT:
                break
    except StopIteration:
        pass
    result+=table_end_template
    return result

ip.display_formatter.formatters['text/html'].for_type_by_name('tables.tableextension',
                                 'Row', rowIterator2Table)
```

###psychopy.iohub.datastore.util.EventAttributeResults objects

```python
def eventQueryResults2Table(event_namedtuple):
    table_start_template="""
        <table align="left" border="1" cellpadding="1" cellspacing="1" style="width: 80%;">
        <tbody>
            <tr>
                <th colspan="1">
                    Condition Variable Values </th>
                <td colspan="{second_col_span}">
                <ul>
                {cv_data}
				</ul>
				</td>
			</tr>
            <tr>
                <th colspan="1">
                    Query String</th>
                <td colspan="{second_col_span}">
                    <pre style="margin-top: 0px; margin-bottom: 0px; padding: 0px; border: 0px; font-family: monospace, sans-serif; font-size: 14px; line-height: 17px; vertical-align: baseline; white-space: pre-wrap; background-color: rgb(255, 255, 255);">
{query_string}
                    </pre>
				</td>
			</tr>
            <tr>
                <th colspan="{num_event_attributes}" style="text-align: center;">
                    Event Matches</th>
            </tr>
				<tr>
                {header_cells}
				</tr>
            """
    header_cell_template="""
            <th>{cell_value}</th>
            """

    event_row_template="""
				<tr>
                {event_cells}
				</tr>
            """

    event_cell_template="""
                <td>{event_value}</td>
            """

    table_end_template="""
            </tbody>
            </table>
            """
    
    cv_data=''
    num_event_attributes=0
    for ck,cv in event_namedtuple.condition_set._asdict().iteritems():
        cv_data+="<li>{ck} : {cv}</li>".format(ck=ck,cv=cv)
        num_event_attributes+=1
    second_col_span=num_event_attributes-1
    query_string='&amp;\n'.join(event_namedtuple.query_string.split('&'))
    query_string=query_string.replace('>','&gt;').replace('<','&lt;')

    ea_names=[cv_name for cv_name in event_namedtuple._asdict().keys() if cv_name not in ('query_string','condition_set')]
    header_cells=''.join([header_cell_template.format(cell_value=cv_name) for cv_name in ea_names])
    
    result=table_start_template.format(cv_data=cv_data,query_string=query_string,num_event_attributes=num_event_attributes,header_cells=header_cells,second_col_span=second_col_span)
    
    event_data_columns=[getattr(event_namedtuple,n) for n in ea_names]
    for i in xrange(len(event_data_columns[0])):
        if i > MAX_ROW_INSTANCE_COUNT:
            break
        event_cells=''.join([event_cell_template.format(event_value=event_data_columns[v][i]) for v in xrange(len(ea_names))])
        result+=event_row_template.format(event_cells=event_cells)         
    result+=table_end_template
    return result
    
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.datastore.util',
                                 'EventAttributeResults', eventQueryResults2Table)
```

##ioHub Runtime Event namedtuples

```python
def ioObject2html(emd):
    ntuple_class_template="""
    <span style="font-family:arial,helvetica,sans-serif;">
    <ul>
    <span style="color:#800000;">
    <strong>
            {0}:
    </strong>
    </span>
    """.format(emd.__class__.__name__)
    
    ntuple_fields_template="""
    <ul>
        {attributes}
    """  
    
    ntuple_attribute_template="""
    <li>
        <strong> {key}: </strong> {value}
    </li>
    """

    ntuple_end_template="""
    </ul></ul>
    """  

    result = ntuple_class_template
    attributes=''.join([ntuple_attribute_template.format(key=key,value=value)for key,value in emd._asdict().iteritems()])
    result+=ntuple_fields_template.format(attributes=attributes)
    result+=ntuple_end_template
    return result.replace('\\n','')

ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'KeyboardPressEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'KeyboardReleaseEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'KeyboardCharEventNT', ioObject2html)

ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MouseButtonPressEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MouseButtonReleaseEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MouseMoveEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MouseDragEventNT', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MouseScrollEventNT', ioObject2html)

ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'MonocularEyeSampleEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'BinocularEyeSampleEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'FixationStartEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'FixationEndEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'SaccadeStartEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'SaccadeEndEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'BlinkStartEvent', ioObject2html)
ip.display_formatter.formatters['text/html'].for_type_by_name('psychopy.iohub.devices',
                                 'BlinkEndEvent', ioObject2html)

```
