# High Level ioHub API Review

### Starting the ioHub Server Process

There are two ways to create a psychopy experiment and have it start the iohub process for use during the experiment runtime. Both menthods ultimately give you access to an instance of the [ioHubConnection Class](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/getting_connected.html#the-iohubconnection-class) for communication and control of the iohub framework:

1. Use the [psychopy.iohub.launchHubServer()](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/launchHubServer.html#the-launchhubserver-function] function.)
2. Create a subclass of the [psychopy.iohub.client.ioHubExperimentRuntime](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/ioHubExperimentRuntime.html#the-iohubexperimentruntime-class), implementing the classes run() method for your experiment's starting python code.

Each approach has +'s and -'s:

- Using launchHubServer is quicker when creating an experiment that uses simple devices like the mouse and keyboard. 
- When using more advanced devices like an eye tracker, using the ioHubExperimentRuntime approach quickly becomes easier to manage, as long as you understand the simple file structure expected by the ioHubExperimentRuntime.
** 

#### Using [launchHubServer()](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/launchHubServer.html#the-launchhubserver-function)



```python
from executeNotebook import execute_notebook
execute_notebook("Py4ET_2013_ECEM_Workshop.ipynb")
execute_notebook("ioObjects2html.ipynb")

def startStopHub(**kwargs):
    """
    Starts the ioHub Process using launchHubServer. When started with no key word arguements, 
    the ioHub Process is started with Display, Keyboard, Mouse, and Experiment devices created 
    using their default configurations. 
    The ioDataStore is disabled by default, so no events are saved by the ioHub process. 

    Optional key word arguements:

    session_code: A string providing the code to be assigned to the current run / session of the experiment. 
                  If provided, the ioDataStore is also initialized for event storage.

    experiment_code: A string providing the code to be given to the experiment data stored in the ioDataStore.
                If experiment_code is provided and session_code is not, then session_code is auto generated
                using::
                
                import psychopy.data
                session_code="S_{0}".format(psychopy.data.getDateStr()) 

    psychopy_monitor_name: Specifies the psychopy monitor configuration file to read the the display physical size
                and the eye to display distance to use in ioHub event calculations.

    datastore_name: The name to give the HDF5 file for the experiment if, based on other keyword arguements provided,
                the ioDataStore will be ective during the experiment. A default name is used if needed.

    [iohub_device_class_name]: Any key word arguenet that equals the name of any ioHub Device class, is assumed
                to indicate that you are providing a preferences dictionary containing any non default preference
                values to be used during the experiment.
    """
    io=psychopy.iohub.launchHubServer(**kwargs)

    keyboard=io.devices.keyboard

    display("Press any Key to Stop the ioHub Service:")
    
    keyboard.clearEvents()
    
    while True:
        kb_events=keyboard.getEvents() 
        if kb_events:
            display("Keyboard Event(s) Detected: ")
            for e in kb_events:
                display(e)
            break
        display('.')
        sys.stdout.flush()
        io.wait(0.5)

    io.quit() 

if __name__ == '__main__':
    startStopHub()
    #startStopHub(experiment_code='silly_exp')
    #startStopHub(experiment_code='silly_exp',session_code='s_001_m_42')
    #exp_kwargs={'experiment_code':'silly_exp','session_code'='s_001_m_42','xinput.Gamepad'={}}
    #startStopHub(**exp_kwargs)
    
```

### Extending the [ioHubExperimentRuntime class](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/ioHubExperimentRuntime.html#the-iohubexperimentruntime-class)

When the ioHubExperimentRuntime class is used to enable the ioHub Process for your experiment, the following *experiment file structure* must be used:

- The Python source file containing the ioHubExperimentRuntime subclass for your experiment must be in a directory dedicated to that experiment (lets refer to it as the exp_dir).
- The exp_dir folder must also contain two configuration files:

    - An experiment_config.yaml file specifying experiment and session level settings.
    - An iohub_config.yaml file that mainly provides the configuration information for each ioHub Device that should be enabled for the experiment.

```python
class ExperimentRuntime(psychopy.iohub.ioHubExperimentRuntime):
    """
    Create an experiment using psychopy and the ioHub framework by extending 
    the ioHubExperimentRuntime class. At minimum all that is needed is to override
    the run() method with the main experiment script logic.
    """
    def run(self,*args):
        """
        The run method contains your experiment logic. 
        It is equal to what would be in your main psychopy experiment
        script .py file in a standard psychopy experiment setup. 
        """

        # Clear all events from the global and device level event buffers.
        self.hub.clearEvents('all')
        display("End Demo By Pressing the Space Key:")
        run_demo=True
        while run_demo:
            for k in self.devices.keyboard.getEvents(EventConstants.KEYBOARD_PRESS):
                if k.key == u' ':
                    run_demo=False
                    break
                display(k)
            self.hub.wait(0.25)
        ### End of experiment logic

###############################################################################

if __name__ == "__main__":
    def main(configurationDirectory):
        """
        Creates an instance of the ExperimentRuntime class, 
        and launches the experiment logic in run().
        """
        runtime=ExperimentRuntime(configurationDirectory, "experiment_config.yaml")    
        runtime.start(sys.argv)
    # run the main function, which starts the experiment runtime
    main(module_directory(main))
```

### Devices

In ioHub, a _Device can be Physical or Virtual_. Physical Devices represent a piece of measurement hardware that generates events. Virtual Devices are purely software based, but can also generate events of interest.

Examples of *Physical* ioHub Devices:

- Keyboard
- Mouse
- Gamepad
- Eye Tracker

Examples of *Virtual* ioHub Devices:

- Experiment
- Event Publisher
- Remote Event Subscriber

A list of all currently supported device types, and associated event types each device can generate, can be found [here](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/devices.html#available-iohub-device-and-deviceevent-types). 

As seen in earlier code, iohub devices are accessed from within the psychopy experiment script via the [ioHubConnection](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/getting_connected.html#the-iohubconnection-class) class .devices attribute. Notes:

- The .devices attribute is dynamically created when the experiment starts. 
- The .devices attribute is itself a class, providing *dot name* access to the iohub devices that were enabled for the experiment. Devices that are not enabled for the experiment will not appear in the devices list.
- The devices attribute name for a device is equal to the device label specified in the device configuation file (or the default name is used).
- To programatically check if a device has been enabled from within the experiment script, the deviceByLabel dict attribute of the ioHubConnection class can be used. This is useful so that an experiment can conditionally call a specific device method only if the device is actually enabled.
- The list of methods that can be called for each device can be accessed within the experiment script.

###Device Access Examples

```python
def deviceAccessTest(**kwargs):
    io=psychopy.iohub.launchHubServer(**kwargs)

    display("** List of Available ioHub Devices:")
    
    # Display each available iohub device name and associated device class 
    for device_name,iodevice in io.deviceByLabel.iteritems():
        display("{device_name} : {iodevice}".format(device_name=device_name,iodevice=iodevice))
        
    # Check if a device with a given name is available:
    
    keyboard=io.deviceByLabel.get('keyboard')
    analog_input=io.deviceByLabel.get('ain')
    
    display('** Checking for device availability:')
    display("Keyboard Device Enabled (should == True): {0})".format(keyboard!=None))
    display("Analog Input Device Enabled (should == False): {0})".format(analog_input!=None))

    # It is often useful to *get rid of the dots* for devices that are used frequently
    # through out the experiment scipt.
    
    keyboard = io.devices.keyboard
    mouse = io.devices.mouse
    
    display("** Checking device instance equality:")
    display("keyboard == io.deviceByLabel.get('keyboard'): {0}".format(keyboard == io.deviceByLabel.get('keyboard')))
    display("mouse == io.devices.mouse): {0}".format(mouse == io.devices.mouse))
    io.quit() 

if __name__ == "__main__":
    deviceAccessTest()      
```

Notice that the device classes handed to you by ioHubConnection are of type psychopy.iohub.client.ioHubDeviceView. This is the PsychoPy  Processes representation of the actual ioHub Device class instances running on the seperate ioHub Server Process.

Notes:

- For each enabled ioHub Device running during the experiment, their is:

    - One psychopy.iohub.client.ioHubDeviceView instance created by the Experiment Process
    - One iohub.devices.[DeviceClassName] instance created in the ioHub Server Process

- Any time an the experiment script calls a method of an ioHub Device:

    1. An IPC call is made to the associated ioHub Device on the ioHub Server Process (the request)
    2. Then the ioHub Process sends back the requested information to the Experiment Process (the response)
    
- While the IPC request-response delay is very short ( < 0.5 msec on average ), it is still important to keep it in mind. When possible adopt a *fetch and reuse* approach to reading device values repeatedly within a display refresh cycle. For example:

        def bad_iohub_code():
            """
            Silly sudo code example of repeatedly calling a method 
            of an ioHub Device when not necessary. 
            **BAD** : Would take approx. getPosition() call delay * 5 (msec) to execute.
            Assuming a 0.25 msec average delay, that equals 1.25 msec!
            """
            mouse=io.devices.mouse
            stim.setPos(mouse.getPosition())
            log("Mouse Pos: {}".format(mouse.getPosition())
            print "Mouse Pos: ",mouse.getPosition()
            if last_mouse_pos == mouse.getPosition():
                pass
            else:
                last_mouse_pos= mouse.getPosition()
        
        def good_iohub_code():
            """
            Vs. Using a Fetch and Reuse approach.
            **GOOD** : Would take approx. getPosition() call delay * 1 (msec) to execute.
            Assuming a 0.25 msec average delay, that equals 0.25 msec, or 5 times faster than the above.
            """
            mouse=io.devices.mouse
            current_mouse_pos=mouse.getPosition()
            stim.setPos(current_mouse_pos)
            log("Mouse Pos: {}".format(current_mouse_pos)
            print "Mouse Pos: ",current_mouse_pos
            if last_mouse_pos == current_mouse_pos:
                pass
            else:
                last_mouse_pos = current_mouse_pos

### Events

Each ioHub Device supports 0 - N ioHub Device Events. When a device is being monitored by the ioHub Process, any events that occur:

- Are converted to a common device type independent event field set + a common device type dependent field set. The event attributes for a given device type are the same regardless of the hardware being used to implement the ioHub Device. This is common sense for devices like the keyboard and mouse (at least within a given OS), but is normally far from the case when using devices like eye trackers or DAQs from different manufacturer's.
- Are either time stamped by the ioHub process if the event does not have a device time stamp, or has the device time stamp converted to the psychopy / iohub time base.
- In the case of device time to iohub time base conversion, when possible:

    - Both offset and drift between the two time bases are factored into the conversion.
    - The delay of the event (iohub event received time - device event time) is subtraced from the iohub event time for the device.
    
- Are provided to the PsychoPy Experiment Runtime in real-time when requested.
- Are stored in the ioDataStore; a HDF5 file that uses an event storage structure which constains the same event attributes for each event type as are used by the online accessable event objects during the experiment runtime. 

#### Runtime Event Access:

Events can be retrieved from the ioHub Process from within an experiment script by either:

- Using the ioHubConnection.getEvents() method.
- Using the devices getEvent() method.

Events can be *cleared* from the ioHub Server event buffers using the ioHubConnection.clearEvents() method. Clearing events removes already collected, but not yet requested, events from the ioHub Process buffers so that they not be returned the next time getEvents is called.

Events can be retrived by the experiment process in one of three formats:

- a tuple of the event attribute values, ordered in the same fashion as the attributes are defined for the event type.
- a dict, where keys are the event attribute names, and values are ... , well the event attribute values.
- a namedtuple, allowing access to event attributes as if the evnt was a tuple, or using dot name notation to access the event attributes. i.e. event.time returns the same value as event[TIME_FIELD_INDEX].

By default, events are returned as namedtuples.

#### ioHubConnection.getEvents()

- Returns *all* events from all enabled iohub devices that have been received since the last call to ioHubConnection.getEvents() or ioHubConnection.clearEvents(). So note that calling getEvents() not only returns the events, but removes them from the device type independent event buffer (called the global event buffer).
- Events are returned sorted by the ioHub time of each event.
- Events for a specific device type can be requested from the global event buffer by passing the appropriate device name instance string as an arg to the method. In this case, only events from the given device instance are returned and removed from the global buffer.

#### [device_name].getEvents()

- Returns events from the device_name device type only, which have been received since the last call to device_name.getEvents() or device_name.clearEvents(). So note that calling device_name.getEvents() not only returns that device's events, but removes them from the device instance dependent event buffer (called a device event buffer).
- Events are returned sorted by the ioHub time of each event.
- Events of a specific event type can be requested by providing the desired event type constant to the device_name.getEvents() method.In this case, only events of the requested event type are returned and removed from the device's event buffer.

**Examples:**

```python
def eventManagementTest(**kwargs):
    io=psychopy.iohub.launchHubServer(**kwargs)

    mouse=io.devices.mouse
    keyboard=io.devices.keyboard

    display("** Move the mouse and press some keys now...")
    iohub_events=[]
    mouse_events=[]
    kb_events=[]
    start_time=Computer.getTime()
    while Computer.getTime()-start_time<=2.0:
        display("Collecting events...")      
        iohub_events.extend(io.getEvents())
        mouse_events.extend(mouse.getEvents())
        kb_events.extend(keyboard.getEvents())
        io.wait(0.25)
    display("** Events Collected Count: io.getEvents={0}, mouse.getEvents()={1}, keyboard.getEvents()={2}".format(
                                            len(iohub_events),len(mouse_events),len(kb_events)))
    
    io.quit() 

def eventManagementTest2(**kwargs):
    io=psychopy.iohub.launchHubServer(**kwargs)

    # clear all event buffers (Global and all Device buffers)
    # 
    io.clearEvents('all')

    display("** Get Only Keyboard Events from io.getEvents(): ")
    start_time=Computer.getTime()
    while Computer.getTime()-start_time<=2.0:
        display("Collecting kb events...")    
        kb_events=io.getEvents('keyboard')
        for ke in kb_events:
            display(ke)
        io.wait(0.25)

    io.quit() 

def eventManagementTest3(**kwargs):
    io=psychopy.iohub.launchHubServer(**kwargs)

    mouse=io.devices.mouse

    # clear all event buffers (Global and all Device buffers)
    # 
    io.clearEvents('all')

    display("** Get Only Mouse Button Press Events from mouse.getEvents(): ")
    start_time=Computer.getTime()
    while Computer.getTime()-start_time<=2.0:
        display("Collecting mouse press events...")
        m_events=mouse.getEvents(event_type_id=EventConstants.MOUSE_BUTTON_PRESS)
        for me in m_events:
            display(me)
        io.wait(0.25)

    io.quit() 
```
