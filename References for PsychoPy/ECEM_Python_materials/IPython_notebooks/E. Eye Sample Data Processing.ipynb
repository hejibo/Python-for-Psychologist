#2. Calculating Eye Sample Noise

```python

time=tsamples.time
xpix=getattr(tsamples,gaze_fields[0]).copy()
ypix=getattr(tsamples,gaze_fields[1]).copy()
pupil_trace=getattr(tsamples,pupil_size_field).copy()
valid_data_periods=processSampleEventGaps(xpix,ypix,pupil_trace,invalid_data_mask,'linear')

xdeg,ydeg=vac.pix2deg(xpix,ypix)
xvel=calculateVelocity(time,xdeg)
yvel=calculateVelocity(time,ydeg)

velocity=np.sqrt(xvel*xvel+yvel*yvel)
velocity_trimmed=np.absolute(velocity[~invalid_data_mask[1:-1]])


# filter 
max_velocity_thresh=scipy.stats.scoreatpercentile(velocity_trimmed,50)
keep_all=velocity<=max_velocity_thresh

# calculate STDEV and RMS based sample Noise
notmasked_contiguous=np.ma.extras.notmasked_contiguous
marray=np.ma.array
x_invalid_data_periods=notmasked_contiguous(marray(xdeg[1:-1], mask=keep_all))
for data_slice in x_invalid_data_periods:
    if np.any(pupil_trace[data_slice.start:data_slice.stop][pupil_trace[data_slice.start:data_slice.stop]==0]):
        pass
    elif data_slice.stop-data_slice.start<=2:
        keep_all[data_slice.start:data_slice.stop]=True
x_valid_data_periods=notmasked_contiguous(marray(xdeg[1:-1], mask=keep_all))

stdev_x=0
stdev_y=0
rms_x=0
rms_y=0
for data_slice in x_valid_data_periods:
    region_length=data_slice.stop-data_slice.start
    
    if np.any(pupil_trace[data_slice.start:data_slice.stop][pupil_trace[data_slice.start:data_slice.stop]==0]):
        pass
    elif region_length==1:
        #print keep_left_all[data_slice.start]
        keep_all[data_slice.start]=False
    else:
        stdev_x+=xdeg[data_slice.start:data_slice.stop].std()
        stdev_y+=ydeg[data_slice.start:data_slice.stop].std()
        
        deltas_x=xdeg[data_slice.start+1:data_slice.stop]-xdeg[data_slice.start:data_slice.stop-1]
        deltas_y=ydeg[data_slice.start+1:data_slice.stop]-ydeg[data_slice.start:data_slice.stop-1]
        rms_x+=np.sqrt(np.sum(deltas_x*deltas_x)/len(deltas_x))
        rms_y+=np.sqrt(np.sum(deltas_y*deltas_y)/len(deltas_y))

stdev_x=stdev_x/len(x_valid_data_periods)
stdev_y=stdev_y/len(x_valid_data_periods)
rms_x=rms_x/len(x_valid_data_periods)
rms_y=rms_x/len(x_valid_data_periods)

copy_pupil=pupil_trace.copy()
copy_pupil[~keep_all]=0
velocity[invalid_data_mask[1:-1]]=np.NaN
velocity_cleared=velocity.copy()
velocity_cleared[~keep_all]=np.NaN

fig = plt.figure()
ax = fig.add_subplot(111)
scatter(xdeg[keep_all],ydeg[keep_all],s=10, marker='o', c='r')
ax.set_xlabel("Horizontal Angle")
ax.set_ylabel("Verical Angle")
plt.title("STDEV and RMS Based Eye Sample Noise Calculation\n(stdev: x={0},y={1}; RMS: x={2},y={3} ).".format(stdev_x,stdev_y,rms_x,rms_y)) 
ax.grid(True)
plt.show()
```

#3. Eye Sample Filtering for Spatial Noise Reduction

## Purposes:
1. Reduction of high frequency noise from eye position data.
2. Decreasing velocity and accelleration noise in the eye signal to aid in improved Eye Event Detection (Saccades, Fixations, Blinks).

## Notes:
1. The requirements of the experimental paradigm being used, and the need for real-time eye data signals, will often influence which Filtering alogorithms can be used for a particular study.
2. There is usually a trade-off between the number of data points required by a filtering algorithm / the computational complexity in performing the filter and the quality of the filtered output.
3. An ideal filter would:
    1. require few sample points per calculation, or require only previously collected data points when filting the current point.
    2. be computationally inexpensive (i.e. fast to calculate)
    3. maintain real changes in eye signal positions, throughout the full biological oculomotor frequency range, while removing artificial signal fluctutions and artifacts.
4. I'm not aware of any filter that is /ideal/.

```python
# Some constants used across many of the example filters used here
#

#### USER PROVIDED CONSTANTS ####

# msec sampling interval of eye tracker
#
ET_SAMPLING_HZ=ET_WORKBOOK_INFO['sampling_rate']

# The desired order of the filter to be used
# (when applicable)...
#
FILTER_ORDER=2

# Desired filtering window size in msec
# Specifying the filter size in msec allows
# for varying filter sample window size
# based on the ET sampling rate.
# > ET sampling rate == > filter sample window length
# for a given filter window duration.
# ( used when applicable )
#
FILT_WINDOW_MSEC=20.0

#### CALCULATED CONSTANTS ####

# Calculate filter window size in samples.
# ( used when applicable )
#
FILT_WIN_SAMPLE_CNT=int(FILT_WINDOW_MSEC//(1000.0/ET_SAMPLING_HZ))

# Ensure window sample size is odd.
#
if FILT_WIN_SAMPLE_CNT%2==0:
   FILT_WIN_SAMPLE_CNT+=1 

# Set sigma used in Gaussian Filter to == FILT_WIN_SAMPLE_CNT//3.5
#
SIGMA=FILT_WIN_SAMPLE_CNT//3.5

display('** Filter Sample Window Size: {0}'.format(FILT_WIN_SAMPLE_CNT))
```

```python
# Cutoff Frequency, calculated based on the
# specified percentage of the 
# Nyquist frequency (ET_SAMPLING_HZ/2.0)
# 
NYQUIST_FREQ_PERC=95.0
TWO_PHASE=True

Wn=0.2#((ET_SAMPLING_HZ/2.0)*(NYQUIST_FREQ_PERC*0.01))/ET_SAMPLING_HZ
b, a = scipy.signal.butter(FILTER_ORDER, Wn, 'low')
if TWO_PHASE is True:
    ffunc=scipy.signal.filtfilt
else:
    ffunc=scipy.signal.lfilter
    
time=tsamples.time
xpix=getattr(tsamples,gaze_fields[0]).copy()
ypix=getattr(tsamples,gaze_fields[1]).copy()
pupil_trace=getattr(tsamples,pupil_size_field).copy()
valid_data_periods=processSampleEventGaps(xpix,ypix,pupil_trace,invalid_data_mask,'linear')

xdeg,ydeg=vac.pix2deg(xpix,ypix)

x_filtered = ffunc(b, a, xdeg)
y_filtered = ffunc(b, a, ydeg)

xdeg[invalid_data_mask]=np.NaN
ydeg[invalid_data_mask]=np.NaN
x_filtered[invalid_data_mask]=np.NaN
y_filtered[invalid_data_mask]=np.NaN  

# Plot the full trial work of data, filtered and unfiltered.
#
subplot_info=dict(Unfiltered=(xdeg,ydeg),Filtered=(x_filtered,y_filtered))

plt.rcParams['figure.figsize']=PLOT_SIZE

tmin=time.min()//1
tmax=time.max()//1+1
plt.figure()
plt.title("Eye Sample Data for Trial %d"%(TRIAL_ID))
ax=plt.gca()

for filter_type, eye_data in subplot_info.iteritems():
    ax.plot(time,eye_data[0],label=filter_type+' X Position')
    ax.plot(time,eye_data[1],label=filter_type+' Y Position')
    ax.set_xlabel('Time')
    ax.set_ylabel('Position (Degrees)')
    
plt.xticks(np.arange(tmin,tmax,0.5),rotation='vertical')

trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.fill_between(time, 0, 1, where=pupil_trace==0, facecolor='DarkRed', alpha=0.5, transform=trans)
ax.set_xlabel('Time')
ax.set_ylabel('Position (Degrees)')
plt.legend(loc=(1.01,.8))
plt.show()


# Create zoomed in plot of just 2 second of data
#
fig = plt.figure()
tmin=time.min()+8.5//1
tmax=tmin+2
within_min_max=(time >= tmin) & (tsamples.time < tmax)
plt.title("Trial %d Eye Samples for Interval %.4f - %.4f"%(TRIAL_ID+1,tmin,tmax))
ax=plt.gca()

for filter_type, eye_data in subplot_info.iteritems():
    ax.plot(time[within_min_max],eye_data[0][within_min_max],label=filter_type+' X Position')
    ax.plot(time[within_min_max],eye_data[1][within_min_max],label=filter_type+' Y Position')
    ax.set_xlabel('Time')
    ax.set_ylabel('Position (Degrees)')
    
plt.xticks(np.arange(tmin,tmax,0.05),rotation='vertical')

trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.fill_between(time[within_min_max], 0, 1, where=pupil_trace[within_min_max]==0, facecolor='DarkRed', alpha=0.5, transform=trans)
ax.set_xlabel('Time')
ax.set_ylabel('Position (Degrees)')
plt.legend(loc=(1.01,.8))
plt.show()
```

#[Savitzky Golay Smoothing Filter](http://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_smoothing_filter)

_Quoting from wikipedia:_
    
The Savitzky–Golay smoothing filter is a filter that essentially performs a local polynomial regression (of degree k) on a series of values 
(of at least k+1 points which are treated as being equally spaced in the series) to determine the smoothed value for each point. 
The main advantage of this approach is that it tends to preserve features of the distribution such as relative maxima, minima and width, 
which are usually 'flattened' by other adjacent averaging techniques (like moving averages, for example).

This filtering algorithm was first described in 1964 by [Abraham Savitzky](http://en.wikipedia.org/wiki/Abraham_Savitzky) 
and [Marcel J. E. Golay](http://en.wikipedia.org/wiki/Marcel_J._E._Golay).

Savitzky and Golay's paper is one of the most widely cited papers in the [journal Analytical Chemistry](http://en.wikipedia.org/wiki/Analytical_Chemistry_(journal)) 
and is classed by that journal as one of its "10 seminal papers" saying "it can be argued that the dawn of the computer-controlled analytical instrument can be traced to this article"

**[SciPy / Numpy implementation from the SciPy Cookbook](http://www.scipy.org/Cookbook/SavitzkyGolay)**

#4b. Accelleration Calculation

```python
time=tsamples.time
xpix=getattr(tsamples,gaze_fields[0]).copy()
ypix=getattr(tsamples,gaze_fields[1]).copy()
pupil_trace=getattr(tsamples,pupil_size_field).copy()
valid_data_periods=processSampleEventGaps(xpix,ypix,pupil_trace,invalid_data_mask,'linear')

xdeg,ydeg=vac.pix2deg(xpix,ypix)
xacc=calculateAccelleration(time,xdeg)
yacc=calculateAccelleration(time,ydeg)

accelleration=np.sqrt(xacc*xacc+yacc*yacc)

weights=np.asarray([.5,2,4.,2,.5])
weights=weights/np.sum(weights)
    
ax_filtered = calculateAccelleration(time[2:-2],np.convolve(xdeg, weights,'valid'))
ay_filtered = calculateAccelleration(time[2:-2],np.convolve(ydeg, weights,'valid'))
accelleration_filtered=np.sqrt(ax_filtered*ax_filtered+ay_filtered*ay_filtered)

accelleration[invalid_data_mask[2:-2]]=np.NaN
accelleration_filtered[invalid_data_mask[4:-4]]=np.NaN

plt.rcParams['figure.figsize']=PLOT_SIZE
tmin=time.min()//1
tmax=time.max()//1+1
plt.figure()
plt.title("Filtered vs. Unfiltered Accelleration Data for Trial %d"%(TRIAL_ID))
ax=plt.gca()

ax.plot(time[2:-2],accelleration,label='Unfiltered Accelleration')
ax.plot(time[4:-4],accelleration_filtered,label='Filtered Accelleration')
ax.set_xlabel('Time')
ax.set_ylabel('Accelleration (Degrees/Sec/Sec)')
plt.xticks(np.arange(tmin,tmax,0.5),rotation='vertical')
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.fill_between(time, 0, 1, where=pupil_trace==0, facecolor='DarkRed', alpha=0.5, transform=trans)
plt.legend(loc=(1.01,.8))
plt.show()

plt.figure()
plt.title("Filtered vs. Unfiltered Accelleration Data; clipped between 0 - 10,000 degrees/sec/sec")
ax=plt.gca()
ax.plot(time[2:-2],accelleration,label='Unfiltered Accelleration')
ax.plot(time[4:-4],accelleration_filtered,label='Filtered Accelleration')
ax.set_xlabel('Time')
ax.set_ylabel('Accelleration (Degrees/Sec/Sec)')
plt.xticks(np.arange(tmin,tmax,0.5),rotation='vertical')
plt.ylim(0, 10000)
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
ax.fill_between(time, 0, 1, where=pupil_trace==0, facecolor='DarkRed', alpha=0.5, transform=trans)
plt.legend(loc=(1.01,.8))
plt.show()
```
